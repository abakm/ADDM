# Data Preprocessing

```python
# Drop Duplicates
df.drop_duplicates(inplace=True)
df
```

```python
df.isnull().sum()  # checking for null values
```

```python
# Apply Label Encoding to sex, smoker and region

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to the 'sex' column
df["sex"] = label_encoder.fit_transform(df["sex"])
df["smoker"] = label_encoder.fit_transform(df["smoker"])
df["region"] = label_encoder.fit_transform(df["region"])

df
```

```python
df.dropna(inplace=True)
df
```

## Model Development

```python
# set Independent and dependent variable
x = df[["age", "sex", "bmi", "children", "smoker", "region"]]
y = df["charges"]

```

```python
# traing sets
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# create the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Prediction
model.predict(x_test)
```

## Evaluate the model's performance

```python
# Calculate mean_sqaured_error
from sklearn.metrics import mean_squared_error, r2_score
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
mse
```

The MSE represents the average of the squared differences between your model's predictions and the actual values. A value of 35493102.61165051 indicates that, on average, the squared error in your predictions is quite large.

Remember that the units of MSE are the squared units of your target variable. So, if you were predicting something like insurance costs in dollars, this MSE would be in dollars squared. This can make direct interpretation of the magnitude a bit challenging.
