###  1.Read the dataset to python environment


```python
import pandas as pd

df = pd.read_csv('diabetes.csv')
df

```

### 2. Display the columns in the dataset.

```python
df.columns
```

###  3. Calculate the mean of each column of the dataset.


```python
for column in df.columns:
  print(f"\nColumn:{column}, Mean: {df[column].mean()} Median: {df[column].median()} Mode: {df[column].mode()}")
```

## 4. Check for the null values present in the dataset

```python
null_counts = df.isnull().sum()
null_counts
```

### 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Univariate Analysis
plt.figure(figsize=(5,3))
plt.hist(df['Pregnancies'], color='red')
plt.xlabel('Pregnancies')
plt.ylabel('Frequency')
plt.title('Glucose Distribution')
plt.show()


# Bivariate Analysis num vs num , Scatterplot
sns.scatterplot(x='Pregnancies', y='SkinThickness', data=df, color='r')
plt.xlabel('Pregnancies')
plt.ylabel('SkinThickness')
plt.title('Pregnancies vs SkinThickness')
plt.show()


# Categorical Analysis
sns.countplot(x='Outcome', data=df, palette='Set2')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.title('Outcome Distribution')
plt.show()

# Bivariate Analysis num vs categorical , Boxplot
sns.boxplot(x='Outcome', y='Glucose', data=df, palette='Set2')
plt.xlabel('Outcome')
plt.ylabel('Glucose')
plt.title('Glucose Distribution by Outcome')
plt.show()



```
